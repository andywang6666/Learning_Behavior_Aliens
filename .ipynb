```python
def get_aliens(image_map_path, features_path, window):
    """
    Returns an array of list of stimuli of each alien. 
    """
    
    image_file = pd.read_excel(image_map_path, index_col=0)
    feature_file = pd.read_excel(features_path)
    
    aliens = []
    for index, row in feature_file.iterrows():
        body_id = str(row['Body'])
        
        # Body
        pos = (0.5, 0) if row['Side'] == 'right' else (-0.5, 0)
        path = image_file.loc['body_' + body_id]['Image Path']
        body = visual.ImageStim(window, image=path, pos=pos)
        
        base_id = body_id + '_'
        
        # Arms
        offset_x = 0.1
        offset_y = -0.05
        pos_arms = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'arms_' + str(row['Arms'])]['Image Path']
        arms = visual.ImageStim(window, image=path, pos=pos_arms)
        
        # Legs
        offset_x = 0.05
        offset_y = -0.15
        pos_legs = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'legs_' + str(row['Legs'])]['Image Path']
        legs = visual.ImageStim(window, image=path, pos=pos_legs)
        
        # Eyes
        offset_x = 0.001
        offset_y = 0.1
        pos_eyes = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'eyes_' + str(row['Eyes'])]['Image Path']
        eyes = visual.ImageStim(window, image=path, pos=pos_eyes)
        
        # Mouth
        offset_x = 0
        offset_y = 0.14
        pos_mouth = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'mouth_' + str(row['Mouth'])]['Image Path']
        mouth = visual.ImageStim(window, image=path, pos=pos_mouth)
        
        # Antenna
        offset_x = 0
        offset_y = 0.2
        pos_antenna = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'antenna_' + str(row['Antenna'])]['Image Path']
        antenna = visual.ImageStim(window, image=path, pos=pos_antenna)
        
        # Tail
        offset_x = 0.05
        offset_y = 0.05
        pos_tail = (pos[0] + offset_x, pos[1] + offset_y)
        path = image_file.loc[base_id + 'tail_' + str(row['Tail'])]['Image Path']
        tail = visual.ImageStim(window, image=path, pos=pos_tail)
        
        stim_list = [body, arms, legs, eyes, mouth, antenna, tail]
        
        aliens.append(stim_list)

    return aliens
```

```python
get_aliens(image_map_path, features_path, window)
```
